Task 4 : Pivote Table

Use pivot tables to summarize data in a
dataset

```python
import pandas as pd 
```

```python
df = pd.read_csv("./House Prices.csv")
```

```python
df.columns
```

```python
df = df[['Id','SalePrice',
'LotFrontage',
'LotArea',
'Neighborhood',
'OverallQual',
'OverallCond',
'YearBuilt',
'TotalBsmtSF',
'GrLivArea',
'GarageCars',
'GarageArea',
'PoolArea', 'YrSold', 'SaleType',
'SaleCondition', ]]
```

```python
df.head()
```

```python
df.columns
```

Price Analysis

```python
df.pivot_table(
    index=['SaleType', 'SaleCondition'],
    values=['SalePrice'],
    aggfunc='mean'
)
```

Number of Sales (Using 'Id'):

```python
df.pivot_table(
    index=['SaleType', 'SaleCondition'],
    values=['Id'],
    aggfunc='count'
)
```

Total Sales Amount (Using 'SalePrice'):

```python
df.pivot_table(
    index=['SaleType', 'SaleCondition'],
    values=['SalePrice'],
    aggfunc='sum'
)
```

Neighborhood vs SalesPrice

```python

pivot_table = pd.pivot_table(df, index='Neighborhood', values='SalePrice', aggfunc=['mean', 'median', 'min', 'max', 'std'])

# Rename the columns for clarity
pivot_table.columns = ['Mean SalePrice', 'Median SalePrice', 'Min SalePrice', 'Max SalePrice', 'Standard Deviation']

pivot_table
```

```python
# Assuming your dataset is named 'df'
# Replace 'df' with the actual name of your DataFrame if it's different

# Select the columns you want to include in the pivot table
columns_to_include = ['Neighborhood', 'OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea',
                      'GarageCars', 'GarageArea', 'PoolArea', 'YrSold', 'SaleType', 'SaleCondition']

# Create a pivot table with the selected columns
df[columns_to_include].pivot_table(index='Neighborhood', aggfunc='mean')
```
